# 3. 한글폰트 사용

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os
plt.style.use('seaborn')

plt.rc('font', family='NanumBarunGothic') 
def change_matplotlib_font(font_download_url):
    FONT_PATH = 'MY_FONT'
    
    font_download_cmd = f"wget {font_download_url} -O {FONT_PATH}.zip"
    unzip_cmd = f"unzip -o {FONT_PATH}.zip -d {FONT_PATH}"
    os.system(font_download_cmd)
    os.system(unzip_cmd)
    
    font_files = fm.findSystemFonts(fontpaths=FONT_PATH)
    for font_file in font_files:
        fm.fontManager.addfont(font_file)

    font_name = fm.FontProperties(fname=font_files[0]).get_name()
    matplotlib.rc('font', family=font_name)
    print("font family: ", plt.rcParams['font.family'])

font_download_url = "https://fonts.google.com/download?family=Noto%20Sans%20KR"
change_matplotlib_font(font_download_url)
```

#16. 그림의 구조

figure는 그림이 그려지는 캔버스나 종이를 뜻하고 axes는 하나의 plot, axis는 가로축이나 세로축을 의미합니다.

# 17. Axes 객체와 subplot 명령
하나의 figure안에 여러 개의 plot을 배열 형태로 보여줘야하는 경우도 많이 있습니다. figure 안에 있는 각각의 플롯은 axes 라고 불리는 객체에 속합니다.

figure안에 axes를 생성하려면 원래는 subplot 명령을 사용해서 axes 객체를 얻어야 합니다. 하지만 plot 명령을 바로 사용해도 자동으로 axes를 생성할 수 있습니다.

subplot 명령은 grid 형태의 axes 객체들을 생성하는데 figure가 행렬(matrix)이고 axes가 행렬의 원소라고 생각하면 됩니다.

subplot 명령은 세 개의 인수를 가집니다. 처음 두개는 원소가 전체 그리드 행령의 모양을 지시하는 숫자이고 세 번째 인수는 그릴 위치를 의미합니다.

```python
x1 = np.arange(1,10)
x2 = np.arange(2,20)
y1 = np.arange(1,10)
y2 = np.arange(2,20)

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1,'yo--')
plt.title('2 x 1 figure')
plt.ylabel('y축 label')

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'r--')
plt.xlabel('x축 label')
plt.ylabel('y축 label')


plt.tight_layout()
plt.show()
```

2 x 2 형태의 네 개의 plot이면 아래와 같이 그릴 수 있습니다. subplot의 인수는 (2,2,1)는 221라는 하나의 숫자로 표시할 수도 있습니다. axes 위치는 위에서 아래로 왼쪽에서 오른쪽으로 카운트 합니다.

```python
np.random.seed(0)

plt.subplot(2,2,1)
plt.plot(np.random.rand(5))
plt.title("첫 번째 그래프")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("두 번째 그래프")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("세 번째 그래프")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("네 번째 그래프")

plt.tight_layout()
```

subplots 명령으로 복수의 Axes 객체를 동시에 생성할 수 있습니다. 이때는 2차원 ndarray 형태로 Axes 객체가 반환됩니다.

```python
fig ,ax = plt.subplots(2,2) # 2 x 2 figure를 그리겠다.
# ax를 이용해서 4개를 그리겠다.
np.random.seed(0)

ax[0,0].plot(np.random.rand(5))
ax[0,0].set_title('첫 번째 그래프')

ax[0,1].plot(np.random.rand(5))
ax[0,1].set_title('두 번째 그래프')

ax[1,0].plot(np.random.rand(5))
ax[1,0].set_title('세 번째 그래프')

ax[1,1].plot(np.random.rand(5))
ax[1,1].set_title('네 번째 그래프')

plt.tight_layout()
plt.show()
```

# 18. Axis 객체와 축 <br>
하나의 axes 객체는 두 개 이상의 axis 객체를 가집니다. axis 객체는 그래프의 가로축, 세로축을 나타내는 객체입니다.

여러가지 플롯을 하나의 axes 객체에 표시할 때 y값의 크기가 달라서 표시하기 힘든 경우가 있습니다. 이때는 twinx 명령으로 복수의 y축을 가진 플롯을 만들 수 있습니다. twinx 명령은 x축을 공유하는 새로운 axes 객체를 만듭니다.

```python
fig ,ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.set_title("2개의 y축 한 개의 figure에서 사용하기")
ax1.plot([11,3,9,5,7,1],'r-',label='ax1')
ax1.set_ylabel('ax1의 y축')
ax1.grid(False)

ax2.plot([20,40,120,80,100,60],'g:',label='ax2')
ax2.set_ylabel('ax2의 y축')
ax2.grid(False)

ax1.set_xlabel('공유하는 x축')
plt.show()


```
