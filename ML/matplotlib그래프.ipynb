# 1. bar chart
x 데이터가 카테고리 값인 경우에는 bar 명령과 barh 명령으로 바 차트 시각화를 할 수 있습니다.<br>
 가로 방향으로 bar 차트를 그리려면 barh 명령을 사용하면 됩니다.


```python
y = [4,7,1]
x = np.arange(3)
xlabel = ['첫번째','두번째','세번째']
plt.title('Bar Chart 그려보기')
plt.bar(x,y)
plt.xticks(x,xlabel)
plt.yticks(sorted(y))
plt.xlabel('xlabel')
plt.ylabel('빈도수')
plt.show()
```

**xerr** 인수나 **yerr** 인수를 사용하면 error bar를 추가할 수 있습니다.<br> 또한 alpha를 사용하면 투명도를 지정할 수 있습니다. <br>alpha는 0~1 사이의 값을 가지며 alpha가 0이면 완전 투명, 1이면 완전 불투명 값을 가집니다.

```python
np.random.seed(0)

people = ['성률','의정','슬기','슬비','건주','민찬']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos,performance,xerr = error,alpha = 0.3)
plt.yticks(y_pos,people)
plt.xlabel("x 라벨")
plt.ylabel("y 라벨")

plt.show()
```

# 2. 파이 차트
카테고리 별 값의 상대적인 비교를 해야 할 때는 pie 명령으로 파이 차트를 그릴 수 있습니다.<br> 파이 차트를 그릴 때는 원의 형태를 유지할 수 있도록 plt.axis('equal') 명령을 실행해야 됩니다.

```python
labels = ['성률','의정','슬기','슬비','건주','민찬'] # 파이차트에서 label들의 이름
sizes = [12.5,15,15,15,10,12.5] # 각 pie 들의 크기 지정
colors = ['gold','blue','red','yellowgreen','lightskyblue','lightcoral'] # 각 pie들의 색깔 지정
explode =(0,0.1,0,0.2,0.3,0.4) # 파이들의 중심으로 부터 멀어지는 거리
plt.title('파이 차트')
plt.pie(sizes,explode = explode,labels = labels,colors = colors,
        autopct='%1.1f%%',shadow = True,startangle=90) #autopct는 숫자 값들의 형식 18.8%와 같이
        # shadow는 그림자를 나타냄
        # startangle은 부채꼴이 그려지는 시작 각도를 설정합니다.
plt.axis('equal')
plt.show()
```

# 4. 스캐터 플롯

2차원 데이터에서 실수 데이터 집합의 상관관계를 살펴보려면 scatter 명령으로 scatter plot을 그릴 수 있습니다. <br>scatter plot의 점 하나의 위치는 데이터 하나의 x,y 값입니다.

```python
np.random.seed(0)
x = np.random.normal(0,1,100)
y = np.random.normal(0,1,100)
plt.title('Scatter plot')
plt.scatter(x,y,color = 'r',label = 'scatter')
plt.xlabel('x축')
plt.ylabel('y축',rotation = 0)
plt.legend(loc = 2)
plt.show()
```

# 5. 버블 차트

scatter plot에서 데이터가 2차원이 아니라 3차원 혹은 4차원인 경우에는 <br>점 하나의 크기, 색깔을 이용해서 다른 데이터의 값을 나타낼 수도 있습니다.<br> 이러한 차트를 버블 차트라고 합니다. 크기는 s인수를 사용하고 색깔은 c 인수를 사용해서 지정합니다.

```python
n = 30
np.random.seed(0)
x = np.random.rand(n)
y1 = np.random.rand(n)
y2 = np.random.rand(n)
y3 = np.pi * (15 * np.random.rand(n)) ** 2
plt.title("Bubble Chart")
plt.scatter(x,y1,c = y2,s = y3,cmap='rainbow')
# cmap을 이용하면 matplotlib에서 미리 정해둔 색깔들을 이용할 수 있습니다.
plt.show()
```

# 6. Imshow

image 데이터처럼 행과 열을 가진 행렬 형태의 2차원 데이터는 imshow을 이용해서 2차원 자료의 크기를 색깔로 표시할 수 있습니다.

```python
from sklearn.datasets import load_digits
# sklearn에 내장되어 있는 데이터를 부르는 코드입니다.

digits = load_digits()
x = digits.images[0]
x
```

```python
plt.title("MNIST DIGITS")
plt.imshow(x,interpolation = 'nearest',cmap = 'summer_r') #interpolation nearest 이용
plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.show()
```

interpolation은 보간법을 뜻하고, 픽셀들의 축 위치 간격을 보정하여 이미지가 자연스러운 모양으로 보일 수 있게 하는 방법입니다.<br> imshow()에서는 16가지(none 제외) 보간법이 있고, 'nearest'는 가장 고해상도인 보간법입니다.

interpolation의 종류: <br>'none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric', 'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos'

```python
methods = [
    None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
    'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
    'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos'
]
fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(x, cmap=plt.cm.bone_r, interpolation=interp_method)
    ax.set_title(interp_method)
plt.show()
```

데이터 수치를 색으로 바꾸는 함수를 colormap이라고 합니다. colormap은 cmap 인수로 지정합니다. 사용할 수 있는 colormap은 plt.cm 속성에 포함되어 있어서 사용하고 싶은 colormap이 있을 때 확인해보면 됩니다.

```python
dir(plt.cm)[:10]
```

```python
fig,ax = plt.subplots(1,4,figsize = (12,3),
                      subplot_kw = {'xticks' : [],'yticks' : []})
ax[0].set_title('plt.cm.Blues')
ax[0].imshow(x,interpolation = 'nearest',cmap = plt.cm.Blues)

ax[1].set_title("plt.cm.Blues_r")
ax[1].imshow(x, interpolation='nearest', cmap=plt.cm.Blues_r)

ax[2].set_title("plt.BrBG")
ax[2].imshow(x, interpolation='nearest', cmap='BrBG')

ax[3].set_title("plt.BrBG_r")
ax[3].imshow(x, interpolation='nearest', cmap='BrBG_r')

plt.show()
```

# 7. 컨투어 플롯

컨투어 플롯은 입력 변수 x,y가 두개이고 출력 변수가 z 하나인 경에는 3차원 자료가 됩니다. 3차원 자료를 시각화 하는 방법은 등고선(contour)을 사용하는 방법이 있습니다. contour, contourf 명령을 사용합니다. contour은 등고선만 표시하고, contourf는 색깔로 표시합니다.

이때 입력 변수 x,y는 그대로 사용할 수 없고 meshgrid 명령으로 그리드 포인트 행렬을 만들어야 합니다.

```python
def f(x,y): # 출력 변수 만들기
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)
    # np.exp는 e 함수입니다.

n = 256
x = np.linspace(-3,3,n) # -3부터 3까지 일정한 간격으로 256개의 값을 만듭니다.
y = np.linspace(-3,3,n)
xx, yy = np.meshgrid(x,y)
zz = f(xx,yy)

plt.title("Contour plots")
plt.contourf(xx,yy,zz,alpha=.75,cmap = 'jet')
plt.contour(xx,yy,zz,colors = 'black')
plt.show()

# 똑같은 값을 하나는 색깔로 하나는 등고선으로 표현한 그래프입니다.
```

#8. 3D 서피스 플롯

3차원 그래프는 등고선 플롯과 달리 Axes3D라는 3차원 전용 axes를 생성해서 입체적으로 표시할 수 있습니다.

plot_wireframe, plot_surface 명령을 사용합니다.

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)
xx, yy = np.meshgrid(x,y)
rr = np.sqrt(xx ** 2 + yy ** 2)
zz = np.sin(rr)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D surface plot")
ax.plot_surface(xx,yy,zz, rstride = 1,cstride = 1,cmap = 'hot')
plt.show()
```
